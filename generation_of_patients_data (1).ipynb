```python
def generer_ECG_moy_data_patient(n_patients):
    l=[]
    d=0
    for i in range(n_patients):
        Simulated_ECG = nk.ecg_simulate(duration=15, sampling_rate=200,noise=0.01, heart_rate=100)
        d=Simulated_ECG.mean()
        l.append(d)
    return l
```

```python
l=generer_ECG_data_patient(150)
```

```python
import pandas as pd
data_patients=pd.DataFrame(l,columns=["ECG_moy_of_patients"])
```

```python
def generer_RSP_moy_data_patient(n):
    lst=[]
    d=0
    for i in range(n):
        rsp_sim = nk.rsp_simulate(duration=15, sampling_rate=200,noise=0.01,respiratory_rate=100, method="sinusoidal")
        d=rsp_sim.mean()
        lst.append(d)
    return lst
```

```python
lst=generer_RSP_data_patient(150)
```

```python
data_patients.insert(1,"RSP_moy_of_patients",lst,True)
```

```python
import pandas as pd
import numpy as np
def generer_body_temp_moy_data_patient(temp_min,temp_max,n):
    lst=[]
    d=0
    for i in range(n):
        rng = np.random.RandomState(1)
        a=(temp_max-temp_min)/n
        body_temp = rng.uniform(temp_min, temp_max, size=3000)
        d=body_temp.mean()
        temp_min+=a
        lst.append(d)
    return lst
```

```python
lst=generer_body_temp_moy_data_patient(95,104,150)
```

```python
from simglucose.simulation.env import T1DSimEnv
from simglucose.controller.basal_bolus_ctrller import BBController
from simglucose.sensor.cgm import CGMSensor
from simglucose.actuator.pump import InsulinPump
from simglucose.patient.t1dpatient import T1DPatient
from simglucose.simulation.scenario_gen import RandomScenario
from simglucose.simulation.scenario import CustomScenario
from simglucose.simulation.sim_engine import SimObj, sim, batch_sim
from datetime import timedelta
from datetime import datetime

```

```python
# specify start_time as the beginning of today
dt = datetime(2022,5,20,22,16,0)
now = datetime.now()
start_time = datetime.combine(dt.date(), datetime.min.time())

def generer_CGM_moy_and_insulin_moy(n_patients):
    lst1=[]
    lst2=[]
    for i in range(n_patients):
        dt = datetime(2022,6,12,20,16,0)
        now = datetime.now()
        start_time = datetime.combine(dt.date(), datetime.min.time())
        path = './results'
        # Create a simulation environment
        patient = T1DPatient.withName('adolescent#00{}'.format(i))
        sensor = CGMSensor.withName('Dexcom', seed=1)
        pump = InsulinPump.withName('Insulet')
        scenario = RandomScenario(start_time=start_time, seed=1)
        env = T1DSimEnv(patient, sensor, pump, scenario)
        # Create a controller
        controller = BBController()
        s1 = SimObj(env,controller, timedelta(days=7), animate=False, path=path)
        results1 = sim(s1)
        data2=pd.DataFrame(results1)
        lst1+=data2['CGM'].mean()
        lst2+=data2['insulin'].mean()
    return lst1,lst2
```
